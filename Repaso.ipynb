```
peso = None 
while (peso == None):
  try:
    peso = float (input('Por favor ingresa tu peso en Kg : '))
  except ValueError:
    print ('El valor ingresado no es valido')
altura = None
while (altura == None):
  try:
    altura = float(input('Por favor ingresa tu altura en metros : '))
  except ValueError:
    print ('El valor ingresado no es valido')

def IMC (peso, altura):
    return round((peso /(altura ** 2)), 3)

print (IMC(peso, altura))
```

```
parrafo = input ('Por favor ingresa un párrafo : ')
print(parrafo)
print ('¿El párrafo finaliza en punto?')
print (parrafo.endswith('.'))
print ('¿Cuantas letras existen en el párrafo?')
print (len (parrafo))
print ('¿Cuantas veces aparece la palabra (Hola)?')
print (parrafo.count('Hola'))
print ('¿Cual es la palabra más larga?')
print (max(parrafo, key = len))
print('Párrafo en mayuscula')
print(parrafo.upper())
print('Párrafo en minuscula')
print (parrafo.casefold())



```

```
import matplotlib.pyplot as plt

meses = ['Junio', 'Julio', 'Agosto', 'Septiembre','Noviembre']
ingresos = [2300000,2497562,4345500,3212120,4521444]
plt.bar (meses,ingresos)
plt.title('Ingresos 2020')
plt.xlabel('Meses')
plt.ylabel('Ingresos')
plt.savefig('Graficoingresos.png')
plt.show()
```

```
import matplotlib.pyplot as plt

elementos = ['Cama', 'Televisor','Beauty','Closet','Escritorio']
ocupacion = [30,5,20,30,15]
explodeList = [0,0,0,0.12,0]
plt.title('Elementos vs Ocupación')
plt.pie(ocupacion, labels = elementos, explode = explodeList, shadow = True)
plt.show()
```
