# Pesos Colombianoa a Dolares

```
peso = None 
while (peso == None):
  try:
    peso = int (input('Por favor ingresa el valor en pesos : '))
  except ValueError:
    print ('El valor ingresado no es valido')
    
def conversion (peso):
  print ('El valor en dolares es : ')
  return round((peso /3800),3)
  
print (conversion(peso))
```

# Párrafo

```
parrafo = input ('Por favor ingresa un párrafo : ')
print(parrafo)
print ('¿El párrafo finaliza en punto?')
print (parrafo.endswith('.'))
print ('¿Cuantas letras existen en el párrafo?')
print (len (parrafo))
print ('¿Cuantas veces aparece la palabra (porque)?')
print (parrafo.count('porque'))
parrafoLista = parrafo.split(' ')
print ('¿Cual es la palabra más larga?')
print (max(parrafoLista, key = len))
print ('¿Cual es la palabra más corta?')
print (min(parrafoLista, key = len))
print('Párrafo en mayuscula')
print(parrafo.upper())
print('Párrafo en minuscula')
print (parrafo.casefold())
```

# Gráfico de barras

```
import matplotlib.pyplot as plt

meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre']
ingresos = [2300000,2497562,4345500,3212120,4521444,2500362,5260300,5000000,3500210,4120000,4600000]
plt.bar (meses,ingresos)
plt.title('Ingresos 2020')
plt.xlabel('Meses')
plt.ylabel('Ingresos')
plt.savefig('Graficoingresos.png')
plt.show()
```

# Gráfico de torta

```
import matplotlib.pyplot as plt

ciudades = ['Cali', 'Medellín','Bogotá','Barranquilla','Cartagena']
ocupacion = [619,382,1775,154,609.1]
explodeList = [0,0,0.12,0,0]
plt.title('Ciudades vs Área Km2')
plt.pie(ocupacion, labels = ciudades, explode = explodeList, shadow = True)
plt.show()
```

# Gráfico ECG

```
import pandas as pd 
```

```
ecg = pd.read_csv('ecg.csv',encoding='UTF-8', header=0, delimiter=';').to_dict()
ecg.keys()
x = list (ecg['muestra'].values())
y = list (ecg['valor'].values())
import matplotlib.pyplot as plt
plt.plot(x,y,"y")
plt.title('Graficando un ECG')
plt.xlabel('Tiempo(ms)')
plt.ylabel ('Voltaje (uV)')
plt.savefig('ecg.png')
plt.show()
```
