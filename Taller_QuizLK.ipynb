# SEÑALES ECG


**La señal Electrocardiográfica (ECG)** *es un registro de la actividad eléctrica cardiaca* en el cual, se realizará una breve reseña de su origen y su formación. 

descarga.jfif

```
import pandas as pd 
```

```
ecg = pd.read_csv('ecg.csv',encoding='UTF-8', header=0, delimiter=';').to_dict()
ecg.keys()
x = list (ecg['muestra'].values())
y = list (ecg['valor'].values())
import matplotlib.pyplot as plt
plt.plot(x,y,"r")
plt.title('Graficando un ECG')
plt.xlabel('Tiempo(ms)')
plt.ylabel ('Voltaje (uV)')
plt.savefig('ecg.png')
plt.show()
```
